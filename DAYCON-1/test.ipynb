```python
import pandas as pd
from collections import Counter
import random
```

```python
# 파일 경로는 실제 위치에 맞게 조정하세요
df = pd.read_csv("train.csv")
df.head()
```

```python
# 각 행에서 정답 순서를 문자열로 조합
df['target_text'] = df[['answer_0', 'answer_1', 'answer_2', 'answer_3']].astype(str).agg(' '.join, axis=1)

# 정답 순열 분포 확인
target_counts = Counter(df['target_text'])

print("🔍 상위 10개 target_text 종류와 빈도수:")
for k, v in target_counts.most_common(10):
    print(f"{k} → {v}번")

print(f"\n✅ 전체 고유 target_text 개수: {len(target_counts)}개")
```

```python
# 랜덤 샘플 5개 확인
sample_df = df.sample(5, random_state=42)

for idx, row in sample_df.iterrows():
    sentences = [row[f'sentence_{i}'] for i in range(4)]
    answers = [row[f'answer_{i}'] for i in range(4)]

    print(f"🟦 Sentences: {sentences}")
    print(f"🟥 Target Order: {answers}")
    print(f"➡ Target Text: {' '.join(map(str, answers))}")
    print("---")
```
